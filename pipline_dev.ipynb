```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import KFold
from sklearn.preprocessing import OneHotEncoder
```

```python
# insert and merge the dataframe
df1 = pd.read_csv('data/Training_data.tbx', sep = '\t')
df2 = pd.read_csv('data/Validation_data.tbx', sep = '\t')
df = df1.append(df2, ignore_index=True)
```

```python
df.head()
```

```python
# extract inputs and targets
X = df.iloc[:, :-2].values
y1 = df.iloc [:, -2].values
y2 = df.iloc [:, -1].values
```

```python
# encode the the targets
label_encoder = LabelEncoder()
gt1 = label_encoder.fit_transform(y1)
gt2 = label_encoder.fit_transform(y2)
label_encoder_name_mapping = dict(zip(label_encoder.classes_,
                                         label_encoder.transform(label_encoder.classes_)))
print("Mapping of Label Encoded Classes", label_encoder_name_mapping, sep="\n")
print("Label Encoded Target Variable", gt2, sep="\n")
```

```python
one_hot_encoder = OneHotEncoder(sparse = False)
one_hot_encoded = one_hot_encoder.fit_transform(gt1)
```

```python
# K-Fold Cross-Validation
kf5 = KFold(n_splits=5, shuffle=False)
for train_index, test_index in kf5.split(list(range(len(X)))):
    print(train_index, test_index)
```
